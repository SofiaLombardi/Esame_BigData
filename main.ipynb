Il dataset analizzato contiene informazioni sui veicoli elettrici a batteria (Battery Electric Vehicles - BEVs) e sui veicoli ibridi plug-in (Plug-in Hybrid Electric Vehicles - PHEVs) attualmente registrati presso il Dipartimento dei Trasporti dello Stato di Washington (Washington State Department of Licensing - DOL). Le colonne del dataset includono, ma non sono limitate a:

    VIN (Vehicle Identification Number): Numero identificativo unico del veicolo.
    
    County: Contea in cui il veicolo è registrato.
    
    City: Città in cui il veicolo è registrato.
    
    State: Stato di registrazione (in questo caso, Washington).
    
    Postal Code: Codice postale della registrazione.
    
    Model Year: Anno del modello del veicolo.
    
    Make: Marca del veicolo.
   
    Model: Modello del veicolo.
   
    Electric Vehicle Type: Tipo di veicolo elettrico (BEV o PHEV).
    
    Clean Alternative Fuel Vehicle (CAFV) Eligibility: 
    
    Idoneità del veicolo come veicolo a carburante alternativo pulito.
    
    Electric Range: Autonomia elettrica del veicolo (in miglia).
    
    Base MSRP: Prezzo base suggerito dal produttore (Manufacturer's Suggested Retail Price).
    
    Legislative District: Distretto legislativo in cui il veicolo è registrato.
    
    DOL Vehicle ID: ID del veicolo assegnato dal Dipartimento dei Trasporti.
    
    Vehicle Location: Posizione del veicolo.
    
    Electric Utility: Fornitore di energia elettrica.
    
    2020 Census Tract: Tratto censuario del 2020.

Importazione delle librerie necessarie e caricamento del dataset

```python

import pandas as pd               # Per la manipolazione dei dati
import numpy as np                # Per operazioni numeriche
import matplotlib.pyplot as plt    # Per la visualizzazione dei dati
import seaborn as sns             # Per la visualizzazione avanzata dei dati

# Caricamento del dataset
df = pd.read_csv('Electric_Vehicle_Population_Data.csv')
```

**Esploro il dataframe**

```python

df.info()
```

**Distribuzione dei Tipi di Veicolo nello stato di Washington**

-Conteggio di BEVs vs. PHEVs: Analizza la distribuzione dei veicoli elettrici a batteria (BEVs) e dei veicoli ibridi plug-in (PHEVs) per vedere quale tipo è più prevalente.

```python
# Conteggio di BEVs e PHEVs
vehicle_type_counts = df['Electric Vehicle Type'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=vehicle_type_counts.index, y=vehicle_type_counts.values)
plt.title('Distribuzione di BEVs e PHEVs nello Stato di Washington')
plt.xlabel('Tipo di Veicolo Elettrico')
plt.ylabel('Numero di Veicoli')
plt.show()
```

Nell'analisi della distribuzione dei veicoli elettrici nello Stato di Washington, emerge che i veicoli elettrici a batteria (BEVs) rappresentano la maggioranza, con quasi 175.000 unità registrate, rispetto ai veicoli ibridi plug-in (PHEVs), che ammontano a circa 50.000.

**Analisi dei prezzi**

-Distrinuzione dei prezzi: Analizza la distribuzione del prezzo base (Base MSRP) per entrambi i tipi di veicolo(BEVs e PHEVs)

-Confronto dei Prezzi Medi: Confronta i prezzi medi di BEVs e PHEVs.

```python
# Confronto dei prezzi medi
average_prices = df.groupby('Electric Vehicle Type')['Base MSRP'].mean()

plt.figure(figsize=(8, 5))
average_prices.plot(kind='bar', color=['blue', 'green'])
plt.title('Prezzo Medio Base di BEVs e PHEVs')
plt.xlabel('Tipo di Veicolo Elettrico')
plt.ylabel('Prezzo Medio Base (USD)')
plt.show()
```

Nell'analisi della distribuzione dei prezzi, si osserva che i veicoli ibridi plug-in (PHEVs) hanno un costo medio di circa 1.400 USD, quasi il doppio rispetto ai veicoli elettrici a batteria (BEVs), che si attestano attorno agli 800 USD.

**Analisi dell'Autonomia Elettrica**

-Distribuzione dell'Autonomia: Analizza la distribuzione dell'autonomia elettrica per BEVs e PHEVs

-Confronto delle Autonomie: Confronta l'autonomia media di BEVs e PHEVs

```python
# Confronto dell'autonomia elettrica
average_range = df.groupby('Electric Vehicle Type')['Electric Range'].mean()

plt.figure(figsize=(8, 5))
average_range.plot(kind='bar', color=['blue', 'green'])
plt.title('Autonomia Elettrica Media di BEVs e PHEVs')
plt.xlabel('Tipo di Veicolo Elettrico')
plt.ylabel('Autonomia Elettrica Media (miglia)')
plt.show()
```

L'autonomia media di quasi 60 miglia per i veicoli elettrici a batteria (BEVs) indica che possono percorrere distanze maggiori con una singola carica rispetto ai veicoli ibridi plug-in (PHEVs), che hanno un'autonomia media di circa 30 miglia. Questa differenza suggerisce che i BEVs sono più adatti per viaggi più lunghi esclusivamente in modalità elettrica, mentre i PHEVs potrebbero richiedere un utilizzo più frequente dei loro motori a combustione interna per percorsi più lunghi.

 Analisi Temporale

   -Tendenze nel Tempo: analizza come il numero di BEVs e PHEVs registrati è cambiato nel tempo.


```python
# Conteggio dei veicoli per anno di modello
vehicles_by_year = df['Model Year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
vehicles_by_year.plot(kind='line', marker='o')
plt.title('Numero di BEVs e PHEVs Registrati nel Tempo')
plt.xlabel('Anno di Modello')
plt.ylabel('Numero di Veicoli')
plt.grid()
plt.show()
```

Il grafico mostra l'andamento delle registrazioni di auto elettriche e ibride plug-in nello stato di Washington dal 2000 al 2025. Si osserva una crescita costante dal 2010 al 2020, culminando in un picco di 60. 000 veicoli registrati nel 2024. Questo indica un forte aumento dell'interesse e dell'adozione di veicoli elettrici e ibridi plug-in negli ultimi anni.
